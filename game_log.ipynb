```python
import pandas as pd
from nba_api.stats.endpoints import teamgamelog
import time
import random
from team_games import get_team_log
from nba_api.stats.endpoints import boxscoreadvancedv2
from team_games import get_team_lineup
```

```python
df = pd.read_csv('league_games.csv')
df.head()
```

```python
df_filtered = df[['SEASON_ID', 'GAME_ID', 'GAME_DATE', 'TEAM_ID', 'TEAM_NAME', 'TEAM_ID_2', 'TEAM_NAME_2']]
```

```python
df_filtered.head()
```

```python
def get_team_WL(row):
    time.sleep(random.uniform(0.5,1))
    df1 = get_team_log(row["TEAM_NAME"])
    df2 = get_team_log(row["TEAM_NAME_2"])
    print(("00" + str(row["GAME_ID"])))
    wl1 = df1.loc[df1['Game_ID'] == ("00" + str(row["GAME_ID"])), ['W', 'L', "W_PCT", "MATCHUP"]].values[0]
    wl2 = df2.loc[df2['Game_ID'] == ("00" + str(row["GAME_ID"])), ['W', 'L', "W_PCT"]].values[0]
    row["W_H"], row["L_H"], row["W_A"], row["L_A"],row["WPCT_H"], row["WPCT_A"], row["MATCHUP"] = wl1[0], wl2[1], wl2[0], wl2[1], wl1[2], wl2[2], wl1[3]
    print(type(row))
    return row
```

```python
df_added = df_filtered.apply(get_team_WL, axis=1)
```
