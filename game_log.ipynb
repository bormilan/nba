```python
import pandas as pd
from nba_api.stats.endpoints import teamgamelog
import time
import random
```

```python
df = pd.read_csv('league_games.csv')
df.head()
```

```python
df_filtered = df[['SEASON_ID', 'GAME_ID', 'GAME_DATE', 'TEAM_ID', 'TEAM_NAME', 'TEAM_ID_2', 'TEAM_NAME_2']]
```

```python
df_filtered.head()
```

```python
def get_team_WL(row):
    time.sleep(random.uniform(0.5,1))
    df = pd.DataFrame(teamgamelog.TeamGameLog(team_id=row["TEAM_ID"]).get_normalized_dict()["TeamGameLog"])
    wl = df.loc[df['Game_ID'] == ("00" + str(row["GAME_ID"])), ['W', 'L']].values[0]
    print(wl)
    row["W"], row["L"] = wl[0], wl[1]
    return row
```

```python
df_added = df_filtered.apply(get_team_WL, axis=1)
```
